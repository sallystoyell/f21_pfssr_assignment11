```python
# The percent of english language learners is very skewed, as you might expect. 
```

```python
import statsmodels.formula.api as smf
lm = smf.ols(formula = "Average_Salary ~ Average_Class_Size", data = school_data).fit()
print(lm.summary())

#Average class size accounts for 0.9% of the variance in salary (R^2 = 0.009; p-value < 0.00)
```

```python
# Problem 4 code here (can split into multiple code chunks if you want)
lm2 = smf.ols(formula = "Average_Salary ~ Average_Class_Size + Perc_Economically_Disadvantaged + Perc_English_Language_Learner", data = school_data).fit()
print(lm2.summary())

#Adjusting for the percent of economically disadvantaged and english language learner students, 
# Average class size no longer significantly accounts for average salry (p-value = 0.552)
```

```python
lm3 = smf.ols(formula = "Average_Salary ~ Average_Class_Size + Perc_Economically_Disadvantaged + Perc_English_Language_Learner + School_Type", data = school_data).fit()
print(lm3.summary())

# Adjusting for the percent of economically disadvantaged and english language learner students and school type, 
# Average class size does not significantly account for average salry (p-value = 0.552). Here, school type is a 
# categorical predictor such that the coefficient and such are comparing public school vs. charter school. I'm not sure
# how it's doing that here though, since all of the Charter School salaries are NAN. We can see this reflected in 
# the warning at the bottom "The smallest eigenvalue is 3.03e-28. This might indicate that there are
# strong multicollinearity problems or that the design matrix is singular." and the fact that the rest of the model
# doesn't change (ex. R^2 = 0.2 in both lm2 and lm3)
```
